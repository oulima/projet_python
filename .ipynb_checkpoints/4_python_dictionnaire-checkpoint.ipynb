# Les dictionnaires

Un dictionnaire en python est une sorte de liste mais au lieu d'utiliser des indexs , on utilise des clés , c'est à dire des characteres qui permettent de retrouver des valeurs.

Par exemple, je prends un dictionnaire français, je recherche un mot (ma clé) et j'obiens une définition (la valeur de la clé)

```python
# créer un dictionnaire
dico = {}
#Ajouter une clé à son dictionnaire
dico['manger']='Avaler pour se nourrir (un aliment solide ou consistant) après avoir mâché.'
```

```python
dico['boire']='Avaler (un liquide)'
```

```python
dico['dormir']="Être dans l'état de sommeil"
```

```python
dico
```

# Faisons un vrai dictionaire

Chargeons un fichier dans lequel est rangé le scrap d'un site internet dictionnaire

```python
mot=[]
definition=[]
f = open( "data/dico.txt", "r" )
for line in f:
    mot.append(line.split(' ')[0])
    definition.append(' '.join(line.split(' ')[1:]))
```

Correction pour les homonymes

```python
f = open( "dico.txt", encoding='UTF8' )
#Liste dans aquelle je range les mots
mot = []
# Liste dans laquelle je range les definition
definition=[]
for line in f:
    tmp=line.split(' ')[0]
    # je fais un test pour eviter d'avoir plusieurs mots (clés) identiques
    if tmp in mot:
        tmp=str(tmp)+'_'+str(tmp)[-1]
    mot.append(tmp)
    definition.append(' '.join(line.split(' ')[1:]))
```

Vous pouvez commencer par verifier que les indexes des deux listes sont synchronisées, c est à dire que l'index de la liste mot est cohérent avec celui de la liste definition.
Verifions la longueur des listes, les elements des listes, ...

```python
len(mot),len(definition)
```

Et maintenant faites moi un beau dictionnaire en manipulant ces deux listes. N'essayez pas de le faire à la main ça serait assez long...
Faites une boucle sur les noms qui seront les clés dico[element de la liste nom]=definition[element de la liste nom].
Ou trouver une autre méthode ...

```python
dico={}
i=0
for elem in mot:
    dico[elem]=definition[i]
    i+=1
dico
```

```python
#dict un dictionnaire 
dico1=dict(zip(mot,definition))
dico1
```
