### Synthese
Pouvez vous rassembler les opérations: chargement/remplissage valeurs manquantes/tracer/sauvegarde réalisées plus haut et les encapsuler dans une fonction.
Elle prend en entree un dataframe et retourne ce meme data frame mais nettoyé.

```python
df=pd.read_csv('data/comptagevelo2017.csv',sep=',',encoding='utf-8',parse_dates=['Date'],dayfirst=True,index_col=['Date']) 

def nett(df):
    df.fillna(method='bfill', inplace=True)
    del df['Unnamed: 1']
    df.resample('M').max().plot()
    return df

nett(df)
```

### Sélectionner uniquement les plaintes concernant le  bruit

Pandas permet une recherche assez agreable d'information sur son support prncipal, le dataframe.

Reprenons le df "plaintes" et cherchons via le champ "Complaint Type" les lignes associées ce type de plainte ('Bruit') ('Noise' en anglais)

Vous pouvez utiliser series.str.contains('mot à chercher')

La premiere ligne identifie une plainte : 'Noise - Street/Sidewalk'
    
Nous allons filtrer ce df sur ce type de plainte

```python
# Je veux affichier plaintes tel que la colonne Complaint Type du df plaintes soient 'Noise - Street/Sidewalk'
df[(df['Complaint Type']=='Noise - Street/Sidewalk')]
```

Malheuresement il existe plusieurs type de plaintes pour le bruit (Noise). Ci dessous une méthode qui va recuperer les types de plaintes dont le mot "Noise" fait parti.

```python
l=[]
for elem in df['Complaint Type']:
    if 'Noise' in elem:
        l.append(elem)
        
l
```

Vous pouvez me mettre cette sequence d'instruction en one line encoding ?

https://wiki.python.org/moin/Powerful%20Python%20One-Liners

Pouvez vous obtenir le meme résultat avec Series.str.contains('Le mot que je cherche')

Le probleme ici, c'est que nous avons plusieurs fois les memes intitulé.
Utilisez la fonction set(liste) pour ne conserver que les valeurs uniques


Conserver uniquement les lignes dont le mot Noise apparait dans la colonne Complaints Type



```python
# Liste des indexs recherchés
ind=[]
for elem in set(l):
    #A l'intérieur, le type d'une colonne est pd.Series ou series
    # les series de pandas sont des tableaux numpy internes. Si vous ajoutez .values à la fin de tout Series, 
    # vous aurez son tableau interne en format numpy
    # nous verrons après ce que c'est
    # Ici je transforme le format de cette series en list pour ma minpulation
    ind=(list(df[df['Complaint Type']== elem].index))+ind
```

```python
# Et voilà !
df.loc[ind]
```

Nous aurions pu également arriver au meme résultat en enchainant les Or ici ecrit avec le symbol '|'
